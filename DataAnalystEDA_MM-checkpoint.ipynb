## relationsheep between salary and revenue

```python
##here I want to use linear regression to see if there is a relationship between  revenue and salary estimate.
##the challenge was the data types conversion. 
```

```python
revenue_ranges = ['Unknown / Non-Applicable', '$100 to $500 million (USD)', '$50 to $100 million (USD)', '$10+ billion (USD)', '-1', '$10 to $25 million (USD)', 
                  '$2 to $5 billion (USD)', '$1 to $5 million (USD)', '$25 to $50 million (USD)', 'Less than $1 million (USD)', '$1 to $2 billion (USD)',
                    '$500 million to $1 billion (USD)', '$5 to $10 million (USD)', '$5 to $10 billion (USD)']
revenue_values = [0, 300, 75, 20000, -1, 17.5, 2500, 3, 37.5, 0.5, 1500, 750, 7.5, 7500]
df_copy["Middle_Revenue"] = df_copy["Revenue"].map(dict(zip(revenue_ranges, revenue_values)))
```

```python
import re
salary_estimate_middle = []
for salary_range in df_copy["Salary Estimate"]:
    salary_values = re.findall('\d+', salary_range)
    if len(salary_values) >= 2:
        lower_bound = int(salary_values[0])
        upper_bound = int(salary_values[1])
        midpoint = (lower_bound + upper_bound) / 2
        salary_estimate_middle.append(midpoint)
    else:
        salary_estimate_middle.append(None)
df_copy['Middle Salary Estimate'] = salary_estimate_middle
```

```python
df_copy[['Middle Salary Estimate', 'Middle_Revenue']].head(5)
```

```python
df_copy.plot.scatter(x='Middle Salary Estimate', y='Middle_Revenue')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(15,6), gridspec_kw={'width_ratios': [1, 2]})

# Plot the first line in the first subplot
ax1.plot(df_copy.sort_values("Middle Salary Estimate").reset_index().index, df_copy.sort_values("Middle Salary Estimate").reset_index()["Middle Salary Estimate"], label="Middle Salary Estimate")
ax1.set_title("Middle Salary Estimate")
ax1.set_xlabel("Index")
ax1.set_ylabel("Value")

# Plot the second line in the second subplot
ax2.plot(df_copy.sort_values("Middle Salary Estimate").reset_index().index, df_copy.sort_values("Middle Salary Estimate")["Middle_Revenue"], label="Middle Revenue")
ax2.set_title("Middle Revenue")
ax2.set_xlabel("Index")
ax2.set_ylabel("Value")

# Set the overall title for the figure
fig.suptitle("Correlation between Middle Salary Estimate and Middle Revenue")

# Show the plot
plt.show()
```
